```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
plt.rcParams['figure.figsize'] = 15, 6
```

```python
df_all = pd.read_csv("final_dataset -2.csv")
df_all.head()
```

```python
df_all["week"] = df_all.Date
```

```python
def find_week(x):
    return datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%a')
df_all.week = df_all.week.apply(find_week)
df_all.head()
```

```python
df_all_dummy = pd.get_dummies(df_all.week,prefix = '',prefix_sep='')
df_all_dummy.head()
```

```python
df_all = df_all.merge(df_all_dummy, right_index = True, left_index  = True, how = 'outer')
df_all.head()
```

```python
df_all.columns
```

```python
df = df_all.copy()
df.head()
```

```python
conditions = df.drop(['Date', 'STATION', 'TIMEINT', 'ENTRIES_hourly', 'Temperature', 'Humidity','week', 'Fri', 'Mon', 'Sat', 'Sun', 'Thu', 'Tue', 'Wed'],axis=1).columns
conditions
```

```python
df['Rain'] =df['Rain'] + df['Heavy Rain'] + df['Light Rain'] + df['Light Freezing Rain']
df['Snow'] =df['Snow'] + df['Heavy Snow'] + df['Light Snow']
df['Cloudy'] = (df['Partly Cloudy'] + df['Scattered Clouds'] + df['Mostly Cloudy']
                + df['Overcast'] + df['Fog'] + df['Haze'] + df['Light Freezing Fog'])

# Get rid of columns that we don't want
df = df.drop([c for c in conditions if c not in ['Rain', 'Snow', 'Cloudy', 'Clear']],axis=1)
df.head()
```

```python
# Model OLS, WLS
import statsmodels.formula.api as smf
```

```python
df.columns
```

```python
lm = smf.ols(formula = 'ENTRIES_hourly ~ Clear + Humidity + Rain + Snow + Temperature + Cloudy', data = df).fit()
print(lm.summary())
```

```python
lm1 = smf.ols(formula = 'ENTRIES_hourly ~ Clear + Humidity + Rain + Snow + Temperature + Fri + Mon + Sat + Sun + Thu + Tue + Wed', data = df).fit()
print(lm1.summary())
```

```python
lm2 = smf.ols(formula = 'ENTRIES_hourly ~ Clear + Humidity + Rain +  Temperature ', data = df).fit()
print(lm2.summary())
```
